```python
import pandas as pd
import glob
import os

# Path where all 10 CSV files are stored
path = r"C:\Users\DELL\OneDrive\Documents\projects-bluestock--Internship\mutual-fund-analysis\data\raw"

# Get all file names
files = glob.glob(path+r"\*.csv")

print("Files found:", len(files))

# Dictionary to store all datasets
datasets = {}

for file in files:
    name = os.path.basename(file).replace(".csv", "")
    df = pd.read_csv(file)
    datasets[name] = df

print(f"\nLoaded: {name}")
print("Shape:", df.shape)

# Example access
# datasets["file1"]
```

```python
for name, df in datasets.items():
    print("\n" + "="*80)
    print(f"DATASET: {name}")

    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")

    print("\nHead:")
    display(df.head())   # better in Jupyter

    print("\nInfo:")
    df.info()
```

```python
import os

# Create output folder if it doesn't exist
output_path = r"C:\Users\DELL\OneDrive\Documents\projects-bluestock--Internship\mutual-fund-analysis\data\processed"
os.makedirs(output_path, exist_ok=True)

for name, df in datasets.items():
    # remove .csv from filename if needed
    file_name = name.replace(".csv", "")

    save_path = os.path.join(output_path, file_name + "_cleaned.csv")

    df.to_csv(save_path, index=False)

    print("Saved:", save_path)
```
