```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("C:/Users/DELL/OneDrive/Documents/projectIT/mutual-fund-analysis/data/raw/07_scheme_performance.csv")
df.head(5)
```

```python
print("Original Shape:", df.shape)
```

```python
# 1. Identify return columns
return_cols = [col for col in df.columns if "return" in col.lower()]

print("\nReturn Columns Found:")
print(return_cols)
```

```python
# 2. Convert returns to numeric
for col in return_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Check missing values after conversion
print("\nMissing values in return columns:")
print(df[return_cols].isnull().sum())
```

```python
# 3. Flag anomalies in returns
for col in return_cols:
    anomaly_col = col + "_anomaly"

    df[anomaly_col] = (
        (df[col] < -100) |   # less than -100%
        (df[col] > 200)      # greater than 200%
    )

# Count anomalies
anomaly_counts = {}

for col in return_cols:
    anomaly_counts[col] = df[col + "_anomaly"].sum()

print("\nAnomaly Counts:")
print(anomaly_counts)
```

```python
# 4. Validate expense ratio
if "expense_ratio" in df.columns:

    df["expense_ratio"] = pd.to_numeric(
        df["expense_ratio"],
        errors="coerce"
    )

    df["expense_ratio_valid"] = (
        (df["expense_ratio"] >= 0.1) &
        (df["expense_ratio"] <= 2.5)
    )

    invalid_expense = (~df["expense_ratio_valid"]).sum()

    print(f"\nInvalid Expense Ratios: {invalid_expense}")

else:
    print("\n'expense_ratio' column not found.")
```
