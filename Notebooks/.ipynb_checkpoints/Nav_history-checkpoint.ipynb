```python
import pandas as pd
df=pd.read_csv("02_nav_history.csv")
df
```

```python
df.head(5)
```

```python
df.info
```

```python
#  Parse date column
df["date"] = pd.to_datetime(df["date"], dayfirst=True, errors="coerce")
```

```python
# Remove invalid dates
df = df.dropna(subset=["date"])
```

```python
# 2. Remove duplicates
df = df.drop_duplicates()
```

```python
# 4. Sort by amfi_code and date
df = df.sort_values(["amfi_code", "date"])
```

```python
# 5. Forward fill missing NAV
df["nav"] = df.groupby("amfi_code")["nav"].ffill()
```

```python
# 6. Validate NAV > 0
invalid_nav = df[df["nav"] <= 0]
print("Invalid NAV Records:", len(invalid_nav))
```

```python
print("Cleaned Shape:", df.shape)
```

```python
print("\nMissing Values:")
print(df.isnull().sum())

```

```python
print("\nSample Data:")
print(df.head())

```

```python
# 8. Save cleaned file
df.to_csv("data/processed/nav_history_clean.csv", index=False)
print("\n Cleaned file saved as:")
print("data/processed/nav_history_clean.csv")
```
